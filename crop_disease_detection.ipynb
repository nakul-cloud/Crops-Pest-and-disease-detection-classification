```python
import kagglehub

path = kagglehub.dataset_download("yusufmurtaza01/plantdoc-object-detection-dataset")

print("Path:", path)
```

```python
DATA_PATH = "/root/.cache/kagglehub/datasets/yusufmurtaza01/plantdoc-object-detection-dataset/versions/1"
```

```python
import os

os.rename(
    f"{path}/dataset.yaml",
    f"{path}/data.yaml"
)
```

```python
import yaml

yaml_path = f"{DATA_PATH}/data.yaml"

with open(yaml_path, "r") as f:
    data = yaml.safe_load(f)

# Fix paths
data["train"] = f"{DATA_PATH}/images/train"
data["val"] = f"{DATA_PATH}/images/val"

# Save back
with open(yaml_path, "w") as f:
    yaml.dump(data, f)

print("data.yaml fixed!")
```

```python
with open(yaml_path, "r") as f:
    print(f.read())
```

```python
from ultralytics import YOLO

model = YOLO("yolo11s.pt")

model.train(
    data=yaml_path,

    epochs=60,
    imgsz=768,   # 🔥 higher resolution

    batch=16,

    optimizer="AdamW",
    lr0=0.0005,   # 🔥 lower LR

    freeze=0,     # 🔥 unfreeze all layers

    weight_decay=0.0005,

    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,

    degrees=15,
    translate=0.1,
    scale=0.5,

    mosaic=1.0,
    mixup=0.2,

    patience=25,

    cache=True,

    name="stage2_finetuned_model"
)
```

```python
import re

# load your file
log_text = open("/train_log.txt").read()

lines = log_text.split("\n")

epochs = []
map50 = []
map5095 = []
precision = []
recall = []

current_epoch = None

for i in range(len(lines)):

    line = lines[i].strip()

    # 🔹 Detect epoch
    if "/60" in line and "Epoch" not in line:
        parts = line.split()
        for p in parts:
            if "/60" in p:
                current_epoch = int(p.split("/")[0])

    # 🔹 Detect metrics row
    if line.startswith("all"):
        parts = line.split()

        # Last 4 values always metrics
        try:
            p = float(parts[-4])
            r = float(parts[-3])
            m50 = float(parts[-2])
            m95 = float(parts[-1])

            if current_epoch is not None:
                epochs.append(current_epoch)
                precision.append(p)
                recall.append(r)
                map50.append(m50)
                map5095.append(m95)

        except:
            continue

print("✅ Extracted:", len(epochs))
```

```python
data = []

for i in range(len(epochs)):
    data.append({
        "epoch": epochs[i],
        "map50": map50[i],
        "map5095": map5095[i],
        "precision": precision[i],
        "recall": recall[i]
    })

import json
with open("training_curve.json", "w") as f:
    json.dump(data, f, indent=2)
```
