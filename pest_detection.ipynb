```python
image_dir = "/content/ip102/JPEGImages"
ann_dir = "/content/ip102/Annotations"

BASE_DIR = "/content/ip102_yolo"

import os
os.makedirs(f"{BASE_DIR}/images/train", exist_ok=True)
os.makedirs(f"{BASE_DIR}/images/val", exist_ok=True)
os.makedirs(f"{BASE_DIR}/labels/train", exist_ok=True)
os.makedirs(f"{BASE_DIR}/labels/val", exist_ok=True)

print("✅ Paths ready")
```

```python
import xml.etree.ElementTree as ET

def convert_xml_to_yolo(xml_file, img_width, img_height, class_map):
    try:
        tree = ET.parse(xml_file)
        root = tree.getroot()
    except:
        return []

    yolo_labels = []

    for obj in root.findall("object"):
        class_name = obj.find("name").text.strip()

        if class_name not in class_map:
            class_map[class_name] = len(class_map)

        class_id = class_map[class_name]

        bbox = obj.find("bndbox")
        xmin = int(bbox.find("xmin").text)
        ymin = int(bbox.find("ymin").text)
        xmax = int(bbox.find("xmax").text)
        ymax = int(bbox.find("ymax").text)

        x_center = ((xmin + xmax) / 2) / img_width
        y_center = ((ymin + ymax) / 2) / img_height
        width = (xmax - xmin) / img_width
        height = (ymax - ymin) / img_height

        yolo_labels.append(f"{class_id} {x_center} {y_center} {width} {height}")

    return yolo_labels
```

```python
import cv2
import shutil

class_map = {}
images = os.listdir(image_dir)

for i, img_name in enumerate(images):

    img_path = os.path.join(image_dir, img_name)
    xml_path = os.path.join(ann_dir, img_name.replace(".jpg", ".xml"))

    if not os.path.exists(xml_path):
        continue

    img = cv2.imread(img_path)
    if img is None:
        continue

    h, w = img.shape[:2]

    yolo_labels = convert_xml_to_yolo(xml_path, w, h, class_map)

    if len(yolo_labels) == 0:
        continue

    split = "val" if i % 5 == 0 else "train"

    shutil.copy(img_path, f"{BASE_DIR}/images/{split}/{img_name}")

    label_path = f"{BASE_DIR}/labels/{split}/{img_name.replace('.jpg','.txt')}"

    with open(label_path, "w") as f:
        f.write("\n".join(yolo_labels))

print("✅ Conversion done")
print("Total classes:", len(class_map))
```

```python
import yaml

names = list(class_map.keys())

data_yaml = {
    "train": f"{BASE_DIR}/images/train",
    "val": f"{BASE_DIR}/images/val",
    "nc": len(names),
    "names": names
}

yaml_path = f"{BASE_DIR}/data.yaml"

with open(yaml_path, "w") as f:
    yaml.dump(data_yaml, f)

print("✅ YAML ready")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")
model.train(
    data=f"{BASE_DIR}/data.yaml",

    # 🔹 training
    epochs=80,
    imgsz=512,
    batch=32,

    optimizer="AdamW",
    lr0=0.003,
    weight_decay=0.0005,

    # 🔥 COLOR AUGMENTATION
    hsv_h=0.02,     # hue variation
    hsv_s=0.8,      # saturation
    hsv_v=0.5,      # brightness

    # 🔥 GEOMETRIC AUGMENTATION
    degrees=10,     # rotation
    translate=0.1,  # shift
    scale=0.5,      # zoom in/out
    shear=2.0,      # slight distortion

    # 🔥 FLIP AUGMENTATION
    fliplr=0.5,
    flipud=0.2,

    # 🔥 ADVANCED AUGMENTATION
    mosaic=1.0,
    mixup=0.2,

    # 🔥 TRAINING STABILITY
    warmup_epochs=3,
    patience=15,

    # 🔥 PERFORMANCE
    cache=True,
    workers=8,

    name="ip102_pest_model_aug"
)
```
